```python
movie_counts = dados.groupby('director')['title'].count().sort_values(ascending=False)

top_5_directors = movie_counts.head(5)


plt.figure(figsize=(10, 6))
plt.bar(top_5_directors.index, top_5_directors.values, color='red')
plt.xlabel('Diretores')
plt.ylabel('Número de Filmes')
plt.title('Top 5 Diretores com Mais Filmes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Este código tem como objetivo gerar um gráfico de barras que mostra os 5 diretores com mais filmes no conjunto de dados fornecido.

```python
# Filtrar apenas filmes
filmes = dados[dados['type'] == 'Movie']

# Contar lançamentos por ano
lancamentos_por_ano = filmes['release_year'].value_counts().sort_values(ascending=False).head(7)

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.bar(lancamentos_por_ano.index, lancamentos_por_ano.values, color='purple')  # Cor roxa

# Adicionar régua
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Configurar rótulos e título
plt.xlabel('Ano de Lançamento')
plt.ylabel('Número de Filmes Lançados')
plt.title('Top 7 "Anos com Mais Lançamentos de Filmes"')
plt.xticks(rotation=45, ha='right')  # Rotacionar rótulos do eixo x
plt.tight_layout()
plt.show()
```

este codigo plota  um gráfico com o top 7 anos que tiveram mais lançamentos de filmes, com a cor roxa e uma régua

```python
# Filtra TV Shows e suas durações
tv_shows = dados[dados['type'] == 'TV Show'].sort_values(by=['duration'], ascending=False)

# separa o  top 5 TV Shows
top_5_tv_shows = tv_shows.head(5)

# extrai as informações de duração e titulos dos shows
duration = top_5_tv_shows['duration'].str.extract('(\d+)').astype(int)  # Extrai numero de duração
titles = top_5_tv_shows['title']

# Cria o plot
plt.figure(figsize=(10, 6))
plt.bar(titles, duration[0], color='orange')  # Use orange color
plt.xlabel('TV Show Title')
plt.ylabel('Duration (Seasons/Episodes)')
plt.title('Top 5 TV Shows with Longest Duration')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

esse codigo plota um grafico da cor laramja com o top 5 series com maior duração

```python
# 1. Crie uma lista de todos os atores da coluna 'cast'
all_actors = []
for cast_list in dados['cast'].dropna():  # Itera pela coluna 'cast', ignorando valores NaN
    atores = str(cast_list).split(',')
    all_actors.extend(atores)

# 2. Remova valores 'nan' da lista
cleaned_actors = [ator for ator in all_actors if ator != 'nan']

# 3. Crie uma série Pandas para contar as ocorrências do ator (usando a lista limpa)
actor_counts = pd.Series(cleaned_actors).value_counts()

# 4. Obtenha os 6 principais atores
top_6_actors = actor_counts.head(6)

# 5. Crie o gráfico
plt.figure(figsize=(10, 6))
plt.bar(top_6_actors.index, top_6_actors.values, color='pink')
plt.xlabel('Atores')
plt.ylabel('Número de Aparições')
plt.title('Top 6 Atores com Mais Aparições')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_6_actors.values, labels=top_6_actors.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Top 6 Atores com Mais Aparições')
plt.axis('equal')  # Garante que o gráfico de pizza seja circular
plt.show()
```

esse grafico plota os 6 atores que mais aparecem na lista e exclui os "nan" valores da lista

```python
# 1. Traduzir e simplificar as classificações
def traduzir_classificacao(rating):
    """Traduz e simplifica as classificações indicativas."""
    if rating in ['TV-MA', 'R', 'NC-17']:
        return 'Adultos'
    elif rating in ['TV-14', 'PG-13']:
        return 'Adolescentes'
    elif rating in ['TV-PG', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-G']:
        return 'Crianças'
    else:
        return 'Outros'

# Aplicar a função para criar uma nova coluna com as classificações traduzidas
dados['classificacao_traduzida'] = dados['rating'].apply(traduzir_classificacao)

# 2. Agrupar por ano e classificação traduzida
rating_distribution = dados.groupby(['release_year', 'classificacao_traduzida'])['title'].count().unstack()

# 3. Criar o gráfico com as classificações traduzidas
rating_distribution.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribuição da Classificação Indicativa ao Longo do Tempo (Traduzido)')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Número de Títulos')
plt.legend(title='Classificação Indicativa')
plt.show()
```

 O código analisa o conjunto de dados dados, agrupa os títulos por ano de lançamento e classificação indicativa, e gera um gráfico de barras empilhadas para visualizar a distribuição dessas classificações ao longo do tempo. Isso permite uma análise da evolução da classificação indicativa dos conteúdos presentes no conjunto de dados.
