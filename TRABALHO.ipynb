```python
movie_counts = dados.groupby('director')['title'].count().sort_values(ascending=False)

top_5_directors = movie_counts.head(5)


plt.figure(figsize=(10, 6))
plt.bar(top_5_directors.index, top_5_directors.values, color='red')
plt.xlabel('Diretores')
plt.ylabel('N√∫mero de Filmes')
plt.title('Top 5 Diretores com Mais Filmes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Este c√≥digo tem como objetivo gerar um gr√°fico de barras que mostra os 5 diretores com mais filmes no conjunto de dados fornecido.

```python
# Filtrar apenas filmes
filmes = dados[dados['type'] == 'Movie']

# Contar lan√ßamentos por ano
lancamentos_por_ano = filmes['release_year'].value_counts().sort_values(ascending=False).head(7)

# Criar o gr√°fico
plt.figure(figsize=(10, 6))
plt.bar(lancamentos_por_ano.index, lancamentos_por_ano.values, color='purple')  # Cor roxa

# Adicionar r√©gua
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Configurar r√≥tulos e t√≠tulo
plt.xlabel('Ano de Lan√ßamento')
plt.ylabel('N√∫mero de Filmes Lan√ßados')
plt.title('Top 7 "Anos com Mais Lan√ßamentos de Filmes"')
plt.xticks(rotation=45, ha='right')  # Rotacionar r√≥tulos do eixo x
plt.tight_layout()
plt.show()
```

este codigo plota  um gr√°fico com o top 7 anos que tiveram mais lan√ßamentos de filmes, com a cor roxa e uma r√©gua

```python
# Filtra TV Shows e suas dura√ß√µes
tv_shows = dados[dados['type'] == 'TV Show'].sort_values(by=['duration'], ascending=False)

# separa o  top 5 TV Shows
top_5_tv_shows = tv_shows.head(5)

# extrai as informa√ß√µes de dura√ß√£o e titulos dos shows
duration = top_5_tv_shows['duration'].str.extract('(\d+)').astype(int)  # Extrai numero de dura√ß√£o
titles = top_5_tv_shows['title']

# Cria o plot
plt.figure(figsize=(10, 6))
plt.bar(titles, duration[0], color='orange')  # Use orange color
plt.xlabel('TV Show Title')
plt.ylabel('Duration (Seasons/Episodes)')
plt.title('Top 5 TV Shows with Longest Duration')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

esse codigo plota um grafico da cor laramja com o top 5 series com maior dura√ß√£o

```python
# 1. Crie uma lista de todos os atores da coluna 'cast'
all_actors = []
for cast_list in dados['cast'].dropna():  # Itera pela coluna 'cast', ignorando valores NaN
    atores = str(cast_list).split(',')
    all_actors.extend(atores)

# 2. Remova valores 'nan' da lista
cleaned_actors = [ator for ator in all_actors if ator != 'nan']

# 3. Crie uma s√©rie Pandas para contar as ocorr√™ncias do ator (usando a lista limpa)
actor_counts = pd.Series(cleaned_actors).value_counts()

# 4. Obtenha os 6 principais atores
top_6_actors = actor_counts.head(6)

# 5. Crie o gr√°fico
plt.figure(figsize=(10, 6))
plt.bar(top_6_actors.index, top_6_actors.values, color='pink')
plt.xlabel('Atores')
plt.ylabel('N√∫mero de Apari√ß√µes')
plt.title('Top 6 Atores com Mais Apari√ß√µes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_6_actors.values, labels=top_6_actors.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Top 6 Atores com Mais Apari√ß√µes')
plt.axis('equal')  # Garante que o gr√°fico de pizza seja circular
plt.show()
```

esse grafico plota os 6 atores que mais aparecem na lista e exclui os "nan" valores da lista

```python
# 1. Traduzir e simplificar as classifica√ß√µes
def traduzir_classificacao(rating):
    """Traduz e simplifica as classifica√ß√µes indicativas."""
    if rating in ['TV-MA', 'R', 'NC-17']:
        return 'Adultos'
    elif rating in ['TV-14', 'PG-13']:
        return 'Adolescentes'
    elif rating in ['TV-PG', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-G']:
        return 'Crian√ßas'
    else:
        return 'Outros'

# Aplicar a fun√ß√£o para criar uma nova coluna com as classifica√ß√µes traduzidas
dados['classificacao_traduzida'] = dados['rating'].apply(traduzir_classificacao)

# 2. Agrupar por ano e classifica√ß√£o traduzida
rating_distribution = dados.groupby(['release_year', 'classificacao_traduzida'])['title'].count().unstack()

# 3. Criar o gr√°fico com as classifica√ß√µes traduzidas
rating_distribution.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribui√ß√£o da Classifica√ß√£o Indicativa ao Longo do Tempo (Traduzido)')
plt.xlabel('Ano de Lan√ßamento')
plt.ylabel('N√∫mero de T√≠tulos')
plt.legend(title='Classifica√ß√£o Indicativa')
plt.show()
```

 O c√≥digo analisa o conjunto de dados dados, agrupa os t√≠tulos por ano de lan√ßamento e classifica√ß√£o indicativa, e gera um gr√°fico de barras empilhadas para visualizar a distribui√ß√£o dessas classifica√ß√µes ao longo do tempo. Isso permite uma an√°lise da evolu√ß√£o da classifica√ß√£o indicativa dos conte√∫dos presentes no conjunto de dados.
